# Корреляция

Корреляция (Пирсона) случайных величин $\xi$ и $\chi$ описывает степень __линейной__ зависимости между данными величинами.

$$
r(\xi, \chi) = \frac{E(\xi\chi) - E(\xi)E(\chi)}{\sigma(\xi)\sigma(\chi)}
$$

Если случайные величины $\xi$ и $\chi$ независимы, то 
$$
E(\xi\chi) = E(\xi) E(\chi).
$$
Следовательно
$$
r(\xi, \chi) = \frac{E(\xi\chi) - E(\xi)E(\chi)}{\sigma(\xi)\sigma(\chi)} = \frac{E(\xi) E(\chi) - E(\xi)E(\chi)}{\sigma(\xi)\sigma(\chi)} = 0
$$

__Обратное неверно!__

Из того, что для некоторых случайных величин верно $r(\xi, \chi) = 0$ не следует, что они независимы:

https://upload.wikimedia.org/wikipedia/commons/0/02/Correlation_examples.png



Корреляция (Пирсона) для выборок $X = (x_1, \dots, x_n)$ и $Y = (Y_1, \dots, Y_n)$ описывает степень __линейной__ зависимости между данными выборками
$$
r(X, Y) = \frac{\frac{1}{n}\sum_{i=1}^{n}(x_i y_i) - \overline{X}\overline{Y}}{\sqrt{(\sum_{i=1}^{n}(x_i - \overline{X}))} \sqrt{(\sum_{i=1}^{n}(y_i - \overline{Y}))}}
$$

```python
n = 1000
X = np.random.random(n)
Y = 2*np.random.random(n)+1
print(np.corrcoef(X,Y))
```

```python
#Для удобства можно определить такую функцию
def corr(X,Y):
    return np.corrcoef(X,Y)[0,1]
```

```python
n = 1000
X = np.random.random(n)
Y = 2*np.random.random(n)+1
print('Коэффициент корреляции = %f' % corr(X,Y))
plt.scatter(X,Y,marker='.',alpha=0.5);
```

```python
n = 1000
X = np.random.random(n)
Y = 2*X+1
print('Коэффициент корреляции = %f' % corr(X,Y))
plt.scatter(X,Y,marker='.',alpha=0.5);
```

```python
n = 1000
X = np.random.random(n)
Y = 20*X+1 # Не зависит от величны коэффициента
print('Коэффициент корреляции = %f' % corr(X,Y)) 
plt.scatter(X,Y,marker='.',alpha=0.5);
```

```python
n = 1000
X = np.random.random(n)
Y = -X+1 # Зависит от знака
print('Коэффициент корреляции = %f' % corr(X,Y))
plt.scatter(X,Y,marker='.',alpha=0.5);
```

```python
n = 1000
X = np.random.random(n)
Error = np.random.normal(0,0.3,n)
Y = 2*X+1 + Error
print('Коэффициент корреляции = %f' % corr(X,Y))
plt.scatter(X,Y,marker='.',alpha=0.5);
```

```python
n = 1000
T = np.random.random(n)*2*math.pi
X = np.cos(T) # Между величинами есть зависимость.
Y = np.sin(T) # Зная X, можно с точносью до знака определить Y.
print('Коэффициент корреляции = %f' % corr(X,Y)) # но корреляция почти равна 0.
plt.scatter(X,Y,marker='.',alpha=0.5);
```

```python
# Продемонстируем, как меняется коэффциент корреляции при увеличении разброса в ошибке
n = 1000
Sigmas = np.linspace(0,10)
Corrs = []
X = np.random.random(n)
for sigma in Sigmas:
    Error = np.random.normal(0,sigma,n)
    Y = X + Error
    Corrs.append(corr(X,Y))
plt.plot(Sigmas, Corrs);
plt.xlabel("Значение отклонения в нормальном рспаределении ошибки")
plt.ylabel("Корреляции");
```
