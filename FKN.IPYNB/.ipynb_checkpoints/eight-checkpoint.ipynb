## Равномерное распределение
Параметры: a,b - концы отрезка
$$
p(x) = 
 \begin{cases}
   \frac{1}{b-a} & x \in [a,b]\\
   0 & x \notin [a,b]\\
 \end{cases}
$$

Математическое ожидание $E(\xi) = \frac{a+b}{2}$

Стандартное отклонение $\sigma(\xi) = \frac{|b-a|}{2\sqrt{3}}$

```python
def uniform_distribution(X,a=0,b=1):
    result = []
    for x in X:
        if a <= x <= b:
            result.append(1/(b-a))
        else:
            result.append(0)
    return np.array(result)

X = np.linspace(-5,5,num=500)
a = -1
b = 3

sample_size = [1000, 10000, 100000]

for n in sample_size:
    Z = np.random.rand(n)*(b-a)+a
    plt.figure()
    plt.hist(Z,density=True, bins=int(n**0.5));
    plt.plot(X,uniform_distribution(X,a,b),color='r')
```

## Задание 1
Смоделируйте __непрерывную__ равномерно распределенную случайную величину на отрезке $[a,b]$ используя только случайную величину, которая равновероятно принимает значения $\{0,1\}$.

Другими словами, Вам надо самостоятельно запрограммировать функцию __np.random.rand(n)*(b-a)+a__, используя только __np.random.randint(0,2)__.

Используйте для решения задачи идею деления отрезка пополам и выбора части отрезка в зависимости от значения __np.random.randint(0,2)__.

```python
def my_uniform(n,a,b):
    result = np.empty(n)
    
    for i in range(n):
        summ = 0
        for j in range(10):
            temp = np.random.randint(0,2)  
            summ += temp /(2 ** (j+1))
        result[i] = (summ) * (b-a) + a
    return result
    
X = np.linspace(-5,5,num=500)
a = -1
b = 3
sample_size = [1000, 10000, 100000]

for n in sample_size:
    Z = my_uniform(n,a,b)
    plt.figure()
    plt.hist(Z,density=True, bins=int(n**0.5));
    plt.plot(X,uniform_distribution(X,a,b),color='r')
```

## Задание 2
Смоделируйте __дискретную__ равномерно распределенную случайную величину, принимающую значения $\{a, a+1, \dots, b\}$,  используя только случайную величину, которая равновероятно принимает значения $\{0,1\}$.

Другими словами, Вам надо самостоятельно запрограммировать функцию __np.random.randint(a,b,size)__ используя только __np.random.randint(0,2)__

```python
def my_uniform(a,b, n):
    result = np.empty(n)
    
    for i in range(n):
        summ = 0
        for j in range(10):
            temp = np.random.randint(0,2)  
            summ += temp /(2 ** (j+1))
        result[i] = (summ) * (b-a) + a
        result[i] = round(result[i])
    return result

print(my_uniform(-4, 5, 6))
```
