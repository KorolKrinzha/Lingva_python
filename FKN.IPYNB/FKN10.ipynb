```python
import numpy as np

import pandas as pd
import random as rnd
from random import shuffle
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def generate_sampling(k,l,sigma,n):
    X = np.empty((2,n))
    X[0] = np.random.normal(0,10,n)
    X[1] = k*X[0] + l + np.random.normal(0,sigma,n)
    df = pd.DataFrame(X.T)
    return df

df = generate_sampling(2,3,10,100)

X_train, X_test, y_train, y_test = train_test_split(df[[0]], df[1], random_state=11)
plt.scatter(X_train, y_train,  c='b', alpha=0.4, label = 'Тренировочная выборка');
plt.scatter(X_test, y_test,  c='r', alpha=0.7, label = 'Тестовая выборка');
plt.legend();
```

```python
def my_linear_reg(X_train, X_test, y_train, y_test):
    regr = LinearRegression()
    regr.fit(X_train, y_train)
    y_train_predict = regr.predict(X_train) 
    y_test_predict = regr.predict(X_test)
    
    print('Коэффициенты прямой: k = %.2f,' % regr.coef_[0], ' l = %.2f' % regr.intercept_)

    print('Среднеквадратическая ошибка (Mean squared error, MSE) на обучающей выборке: %.2f'
          % mean_squared_error(y_train, y_train_predict))
    
    print('Среднеквадратическая ошибка (Mean squared error, MSE) на тестовой выборке: %.2f'
          % mean_squared_error(y_test, y_test_predict))

    print('Коэффициент детерминации на обучающей выборке: %.2f'
          % r2_score(y_train, y_train_predict))

    print('Коэффициент детерминации на тестовой выборке: %.2f'
          % r2_score(y_test, y_test_predict))

    
    plt.scatter(X_train, y_train, c='b', alpha=0.4, label = 'Тренировочная выборка');
    plt.scatter(X_test, y_test, c='r', alpha=0.7, label = 'Тестовая выборка');
    plt.scatter(X_test, y_test_predict, c='g', alpha=0.7, label = 'Предсказания на тесте');
    
    
    X_min = min(X_train.min().item(), X_test.min().item())
    X_max = max(X_train.max().item(), X_test.max().item())
    plt.plot([X_min, X_max], regr.predict(pd.DataFrame([X_min, X_max])),alpha=0.7, label = 'Модель');
    plt.legend()
    
```

```python
sigma = 30
df = generate_sampling(2,3,sigma,100)
X_train, X_test, y_train, y_test = train_test_split(df[[0]], df[1], random_state=11)
my_linear_reg(X_train, X_test, y_train, y_test)
```

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
y_predict = regr.predict(X_test)
y_train_predict = regr.predict(X_train) 
sigma_1 = sum((y_test - y_test.mean())**2)
sigma_2 = sum((y_test - y_predict)**2)
print('Среднеквадратическая ошибка:',sigma_2/len(y_test))
print('Коэффициент детерминации:', 1  - sigma_2/sigma_1)
```
